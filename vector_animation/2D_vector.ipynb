```python
import matplotlib.pyplot as plt
```

```python
v1 = [2, 5]
v2 = [6, 8]
```

```python
plt.quiver(0, 0, v1[0], v1[1], angles="xy", scale_units="xy", scale=1, color="red")
plt.quiver(0, 0, v2[0], v2[1], angles="xy", scale_units="xy", scale=1, color="blue")

plt.xlim([0, 8])
plt.ylim([0, 8])

plt.xlabel("X")
plt.ylabel("Y")

plt.grid()

plt.show()
```

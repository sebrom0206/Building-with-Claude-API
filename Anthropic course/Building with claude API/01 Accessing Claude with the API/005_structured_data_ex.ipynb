````python
# load env variables
from dotenv import load_dotenv

load_dotenv()

# create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"


def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages, system="", temperature=1.0, stop_sequences=[]):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
        system=system,
        temperature=temperature,
        stop_sequences=stop_sequences,
    )
    return message.content[0].text
````

````python
# claude-sonnet-4-6 doesn't support assistant message prefill, so instead of
# prefilling "```" we ask for the fenced block in the prompt and strip it out after.
# threre shouldn't be any comments or explanation

messages = []

prompt = """
Generate three different sample AWS CLI commands. Each should be very short.
Respond with ONLY a single code block wrapped in triple backticks, no other text.
"""

add_user_message(messages, prompt)

text = chat(messages)

# strip the ``` fences from the response
text.strip().strip("`").strip()
````
